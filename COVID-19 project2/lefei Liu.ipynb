# Get the data for The whole world COVID-19

```python
# # Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File paths for the datasets
file_paths = [
    'global_covid_metrics2020.csv',
    'global_covid_metrics2021_H1.csv',
    'global_covid_metrics2021_H2.csv',
    'global_covid_metrics2022_H1.csv',
    'global_covid_metrics2022_H2.csv',
    'global_covid_metrics2023_H1.csv',
    'global_covid_metrics2023_H2.csv'
]

# Load and concatenate all datasets
all_data = pd.concat([pd.read_csv(file) for file in file_paths])

# Display the first few rows of the combined dataset
all_data.head()
```

# Let's continue with the detailed analysis of the combined global COVID metrics dataset

```python


# Grouping by country and calculating total sum for each category
grouped_data = all_data.groupby('Country').sum()

# Sorting and getting top 10 countries for different metrics
top_10_confirmed_cases = grouped_data['Confirmed Cases'].sort_values(ascending=False).head(10)
top_10_confirmed_deaths = grouped_data['Confirmed Deaths'].sort_values(ascending=False).head(10)
top_10_new_cases = grouped_data['New Cases'].sort_values(ascending=False).head(10)
top_10_new_deaths = grouped_data['New Deaths'].sort_values(ascending=False).head(10)

# Plotting the top 10 countries for each metric
plt.figure(figsize=(20, 15))
```

# We can get Top 10 Countries by Total Confirmed Cases. 

```python
# Analysis and Plot for Top 10 Countries by Total Confirmed Cases

# Grouping by Country and summing up the Confirmed Cases
total_cases_by_country = all_data.groupby('Country')['Confirmed Cases'].max()

# Sorting and getting the top 10 countries
top_10_total_cases = total_cases_by_country.sort_values(ascending=False).head(10)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_total_cases.values, y=top_10_total_cases.index, palette="viridis")
plt.title('Top 10 Countries by Total Confirmed Cases')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Country')
plt.show()
```

# View peaks in individual graphs

What can peak data represent?

Answer:


Critical Points of Change: Peaks often indicate moments where the situation has reached a critical level. For example, a peak in 'unvaccinated' status might suggest a time when a large portion of a population was unvaccinated, potentially signaling increased risk for disease spread.


Public Health Policy: Peaks can influence public health policy decisions. If certain policies are followed by a decrease in peaks, they might be considered effective.

```python
import pandas as pd

# Load your data
data = pd.read_csv('status death.csv')

# Convert 'Day' column to datetime
data['Day'] = pd.to_datetime(data['Day'])

# Find the peak (maximum) values for each status
peak_unvaccinated_value = data['unvaccinated'].max()
peak_vaccinated_with_value = data['vaccinated_with'].max()
peak_vaccinated_without_value = data['vaccinated_without'].max()

# Find the corresponding dates for these peak values
peak_unvaccinated_date = data[data['unvaccinated']
                              == peak_unvaccinated_value]['Day'].dt.strftime('%Y-%m-%d').iloc[0]
peak_vaccinated_with_date = data[data['vaccinated_with']
                                == peak_vaccinated_with_value]['Day'].dt.strftime('%Y-%m-%d').iloc[0]
peak_vaccinated_without_date= data[data['vaccinated_without']
                                   == peak_vaccinated_without_value]['Day'].dt.strftime('%Y-%m-%d').iloc[0]

# Print the results
print("Peak Unvaccinated Date:", peak_unvaccinated_date)
print("Peak Vaccinated With Date:", peak_vaccinated_with_date)
print("Peak Vaccinated Without Date:", peak_vaccinated_without_date)
```

# Select a special day in the data as statistics

According to U.S. status death data analysis, I chose 2022-09-24 as a special day because this is the day when the mortality rate after the second vaccination officially started.

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from scipy import stats

# Load the dataset
data = pd.read_csv('status death.csv')

# Convert 'Day' to datetime type
data['Day'] = pd.to_datetime(data['Day'], format='%Y-%m-%d')

# Split the dataset
before_sep24 = data[data['Day'] < pd.Timestamp('2022-09-24')]
after_sep24 = data[data['Day'] >= pd.Timestamp('2022-09-24')]

# Fit a linear model
model_before_sep24 = ols('unvaccinated ~ vaccinated_without', data=before_sep24).fit()
model_after_sep24 = ols('unvaccinated ~ vaccinated_without', data=after_sep24).fit()


```

# Diagnostic plots for the model before Sep 24

```python

# Diagnostic plots for the model before Sep 24
fig = plt.figure(figsize=(12, 8))
sm.graphics.plot_regress_exog(model_before_sep24, 'vaccinated_without', fig=fig)
plt.show()
```

# Diagnostic plots for the model after Sep 24

```python
# Diagnostic plots for the model after Sep 24
fig = plt.figure(figsize=(12, 8))
sm.graphics.plot_regress_exog(model_after_sep24, 'vaccinated_without', fig=fig)
plt.show()
```

# According these two deta we can get what result?

First Set of Plots (For Unvaccinated and Vaccinated Once):

Residuals vs Fitted Plot: There is a noticeable pattern where residuals fan out as the number of vaccinated_without increases. This indicates potential heteroscedasticity and that a simple linear model might not be the best fit.

Normal Q-Q Plot: The points do not closely follow the line, suggesting that the residuals may not be normally distributed.

Scale-Location Plot: The spread of residuals increases with the fitted values, which also suggests heteroscedasticity.

Residuals vs Leverage Plot (CCPR Plot): There do not appear to be any points with high leverage that are unduly influencing the model.

Second Set of Plots (For Unvaccinated and Vaccinated Twice):

Residuals vs Fitted Plot: The residuals show a much more random pattern around the fitted line, suggesting a better fit of the model to the data.

Normal Q-Q Plot: The points follow the line more closely than in the first model, indicating a closer approximation to normal distribution of residuals.

Scale-Location Plot: The residuals are more evenly spread across the range of fitted values, indicating a more constant variance (homoscedasticity).

Residuals vs Leverage Plot (CCPR Plot): the same with the first set, there are no clear points of high leverage.
