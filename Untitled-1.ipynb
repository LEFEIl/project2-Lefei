```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('status death.csv')

# Display the first few rows of the dataset
data.head()
```

# Do a second self-check on your own to make sure no data is missing

```python
# Data Preprocessing and Cleaning

# Checking for missing values
missing_values = data.isnull().sum()

# Converting 'Day' to datetime format for easier handling of temporal data
data['Day'] = pd.to_datetime(data['Day'])

# Replacing NaN values with 0 for 'vaccinated_with' and 'vaccinated_without' columns
# as NaN might indicate no data available or zero status
data['vaccinated_with'] = data['vaccinated_with'].fillna(0)
data['vaccinated_without'] = data['vaccinated_without'].fillna(0)

# Summary of preprocessing steps
preprocessing_summary = {
    "Missing Values Before": missing_values,
    "Missing Values After": data.isnull().sum(),
    "Data Types": data.dtypes
}

preprocessing_summary
```

# Next, move on to exploratory data analysis and visualization. 

This will involve creating visual representations of the data to understand trends, patterns, and distributions.-lefei liu PSU

```python
# Setting the style for the plots
sns.set(style="whitegrid")

# Trend Analysis Over Time for Each Status

# Plotting trends for 'unvaccinated'
plt.figure(figsize=(12, 6))
sns.lineplot(x='Day', y='unvaccinated', data=data, label='Unvaccinated')
plt.title('Trend of Unvaccinated Status Over Time')
plt.xlabel('Date')
plt.ylabel('Unvaccinated Status')
plt.legend()
plt.show()

# Plotting trends for 'vaccinated_with'
plt.figure(figsize=(12, 6))
sns.lineplot(x='Day', y='vaccinated_with', data=data, label='Vaccinated With')
plt.title('Trend of Vaccinated With Status Over Time')
plt.xlabel('Date')
plt.ylabel('Vaccinated With Status')
plt.legend()
plt.show()

# Plotting trends for 'vaccinated_without'
plt.figure(figsize=(12, 6))
sns.lineplot(x='Day', y='vaccinated_without', data=data, label='Vaccinated Without')
plt.title('Trend of Vaccinated Without Status Over Time')
plt.xlabel('Date')
plt.ylabel('Vaccinated Without Status')
plt.legend()
plt.show()
```

# View peaks in individual graphs

What can peak data represent?

Answer:


Critical Points of Change: Peaks often indicate moments where the situation has reached a critical level. For example, a peak in 'unvaccinated' status might suggest a time when a large portion of a population was unvaccinated, potentially signaling increased risk for disease spread.


Public Health Policy: Peaks can influence public health policy decisions. If certain policies are followed by a decrease in peaks, they might be considered effective.

```python
import pandas as pd

# Load your data
data = pd.read_csv('status death.csv')

# Convert 'Day' column to datetime
data['Day'] = pd.to_datetime(data['Day'])

# Find the peak (maximum) values for each status
peak_unvaccinated_value = data['unvaccinated'].max()
peak_vaccinated_with_value = data['vaccinated_with'].max()
peak_vaccinated_without_value = data['vaccinated_without'].max()

# Find the corresponding dates for these peak values
peak_unvaccinated_date = data[data['unvaccinated'] == peak_unvaccinated_value]['Day'].dt.strftime('%Y-%m-%d').iloc[0]
peak_vaccinated_with_date = data[data['vaccinated_with'] == peak_vaccinated_with_value]['Day'].dt.strftime('%Y-%m-%d').iloc[0]
peak_vaccinated_without_date = data[data['vaccinated_without'] == peak_vaccinated_without_value]['Day'].dt.strftime('%Y-%m-%d').iloc[0]

# Print the results
print("Peak Unvaccinated Date:", peak_unvaccinated_date)
print("Peak Vaccinated With Date:", peak_vaccinated_with_date)
print("Peak Vaccinated Without Date:", peak_vaccinated_without_date)
```
